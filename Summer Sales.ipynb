```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
# set the graphs to show in the jupyter notebook
%matplotlib inline
import seaborn as sns
```

# importing the data

```python
df=pd.read_csv('Summer_Sales_Data.csv',encoding='unicode_escape')
```

# drop unrelated/blank columns

```python
df.drop(['Status','unnamed1'], axis=1,inplace=True)
```

```python
# check null values
pd.isnull(df).sum()
```

```python
#drop null values
df.dropna(inplace=True)
```

```python
df.describe()
```

# Exploratory Data Analysis

# Gender

```python
gd=sns.countplot(x='Gender',data=df)
for bars in gd.containers:
    gd.bar_label(bars)
```

```python
Sales_gd=df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values('Amount',ascending=False)
```

```python
sns.barplot(x='Gender',y='Amount',data=Sales_gd)
```

From above graphs we can see that most of the buyers are females and even the purchasing power of females are grater than men.

# Age

```python
ag=sns.countplot(data=df,x='Age Group',hue='Gender')
for bars in ag.containers:
    ag.bar_label(bars)
```

```python
Sales_age=df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values('Amount',ascending=False)
sns.barplot(x='Age Group',y='Amount',data=Sales_age)
```

From above graphs we can see that most of the buyers are of age group between 26-35 yrs female

# State

```python
# total numbers of orders from top 10 states
sales_state=df.groupby(['State'],as_index=False)['Orders'].sum().sort_values('Orders',ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data=sales_state,x='State',y='Orders')
```

```python
# total amount/sales from top 10 states
sales_state=df.groupby(['State'],as_index=False)['Amount'].sum().sort_values('Amount',ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data=sales_state,x='State',y='Amount')
```

From above graphs we can see that unexpectedly most of the orders are from Uttar Pradesh, Maharashtra and Karnataka respectively but total sales/amount is from UP, Maharashtra and then Karnataka.

# Marital Status

```python
ax=sns.countplot(data=df,x='Marital_Status')
sns.set({'figure.figsize':(2,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state=df.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum().sort_values('Amount',ascending=False).head(10)
sns.set(rc={'figure.figsize':(5,5)})
sns.barplot(data=sales_state,x='Marital_Status',y='Amount',hue='Gender')
```

From above graphs we can see that most of the buyers are married(women) and they have high purchasing power

# Occupation

```python
a=sns.countplot(data=df,x='Occupation')
sns.set({'figure.figsize':(20,5)})
for bars in a.containers:
    a.bar_label(bars)
```

```python
sales_occu=df.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values('Amount',ascending=False)
sns.set({'figure.figsize':(20,5)})
sns.barplot(data=sales_occu,x='Occupation',y='Amount')
```

From above graphs we can see that most of the buyers are working in IT, Healthcare and Aviation sector

# Product Category

```python
b=sns.countplot(data=df,x='Product_Category')
sns.set({'figure.figsize':(20,5)})
for bars in b.containers:
    b.bar_label(bars)
```

```python
category=df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values('Amount',ascending=False).head(10)
sns.set({'figure.figsize':(20,5)})
sns.barplot(data=category,x='Product_Category',y='Amount')
```

From above graphs we can see that most of the sold products are from Food, Clothing and Electronics Category

```python
# top 10 most sold products
prd_Id=df.groupby(['Product_ID'],as_index=False)['Orders'].sum().sort_values('Orders',ascending=False).head(10)
sns.set({'figure.figsize':(20,5)})
sns.barplot(data=prd_Id,x='Product_ID',y='Orders')
```

# Conclusion:

Married women age group 26-35 yrs from UP, Maharastra and Karnataka working in IT,Healthcare and Aviation are more likely to buy product from Food, Clothing and Electronics Category.
